##### **Data Analyst Job Market Insights - Visualizations**
This notebook contains visualizations based on SQL query results from an analysis of 2023 job postings.  
It focuses on:  
- The **top 10 highest-paying Data Analyst jobs**  
- The **most in-demand skills for the top 10 high-paying roles**  

The dataset used here is a result of SQL queries performed on the datset, and the visualizations help provide a clearer understanding of salary trends and skill demands.  


```python
# importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load data
job = pd.read_csv('top_paying_job.csv')
skill = pd.read_csv('top_paying_job_skills.csv')
```

 **Top 10 Highest-Paying Data Analyst Jobs**

This bar chart visualizes the top-paying Data Analyst roles based on salary.


```python
# bar chart to display top paying jobs
plt.figure(figsize=(8,6))
sns.barplot(x='salary_year_avg',y='job_title', data=job)

# adding label and title
plt.title('Top 10 Paying Jobs')
plt.xlabel("Average Yearly Salary")
plt.ylabel("Job Title")

plt.show()
```

**Skills for Top 10 Highest-Paying Data Analyst Jobs**

This bar chart visualizes the top-paying Data Analyst roles based on salary.


```python
# count occurences of each skill
skill_counts = skill.groupby('skills').agg(skill_count=("job_id","count")).reset_index()

# get the top ten skills
top_skills = skill_counts.nlargest(10, "skill_count")

print(top_skills)
```

```python
# lets visualize the top skills

# Plot bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x="skill_count", y="skills", data=top_skills, hue="skills", palette="viridis",legend=False)

# Add labels and title
plt.xlabel("Count")
plt.ylabel("Skills")
plt.title("Top 10 Skills of the High Paying Jobs")

# Show the plot
plt.show()
```
